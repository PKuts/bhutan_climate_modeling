## ✅ Data Cleaning Summary

We performed a quality check and removed implausible values from the processed meteorological data for Bhutanese regions. The cleaning criteria were:

- **Tmax > 45°C** → removed as unrealistic
- **Tmin < -15°C** → removed as unrealistic
- **Rainfall > 500 mm/day** → removed as potential errors
- **Relative Humidity (RH) > 100%** → physically invalid

### 🔍 Detected Outliers:

#### Nganglam
- **Tmax**
  - 2017-09-07: 229.0°C ❌
- **RH**
  - 2006-06-07: 140.6%
  - 2007-08-01: 131.6%
  - 2019-02-12: 109.88%

All other regions (`Paro`, `Chamkhar`, `Mongar`, `Bhur`, `Gasa`, etc.) had no outliers based on the thresholds above.

The cleaned data has been saved, overwriting the original `.pkl` files in `../processed_data/`.


```python
import os
import numpy as np
import pandas as pd

data_dir = '../processed_data'

# Define thresholds and comparison rules
thresholds = {
    'Tmax': ('>', 45),
    'Tmin': ('<', -15),
    'Rainfall': ('>', 500),
    'RH': ('>', 100)
}

region_files = [f for f in os.listdir(data_dir) if f.endswith('.pkl')]

for file in region_files:
    region_name = file.replace('.pkl', '')
    file_path = os.path.join(data_dir, file)
    df = pd.read_pickle(file_path)

    cleaned = False  # Track whether we made any changes

    print(f"\n=== {region_name} ===")
    for col, (op, val) in thresholds.items():
        if col in df.columns:
            if op == '>':
                outliers = df[df[col] > val][['Date', col]]
                if not outliers.empty:
                    print(f"\nOutliers in '{col}' (> {val}):")
                    print(outliers)
                    df.loc[df[col] > val, col] = np.nan
                    cleaned = True
            elif op == '<':
                outliers = df[df[col] < val][['Date', col]]
                if not outliers.empty:
                    print(f"\nOutliers in '{col}' (< {val}):")
                    print(outliers)
                    df.loc[df[col] < val, col] = np.nan
                    cleaned = True

    # Save only if any outlier was found and removed
    if cleaned:
        df.to_pickle(file_path)
        print(f"✔ Saved cleaned data for {region_name}")
```
